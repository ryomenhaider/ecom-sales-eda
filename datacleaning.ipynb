```python
df['annual_income'] = pd.to_numeric(df['annual_income'], errors='coerce')
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['annual_income'] = df['annual_income'].fillna(df['annual_income'].median())
df['discount_usage_pct'] = df['discount_usage_pct'].fillna(df['discount_usage_pct'].median())
df['email_open_rate'] = df['email_open_rate'].fillna(df['email_open_rate'].median())
df['satisfaction_score'] = df['satisfaction_score'].fillna(df['satisfaction_score'].median())
df['phone'] = df['phone'].fillna('Unknown')
```

```python
df[['phone', 'join_date', 'preferred_category', 'loyalty_tier']] = df[['phone', 'join_date', 'preferred_category', 'loyalty_tier']].fillna('Unknown')
```

```python
numeric_cols = [
    'age',
    'annual_income', 
    'discount_usage_pct', 
    'email_open_rate', 
    'satisfaction_score', 
    'total_purchases', 
    'total_spend', 
    'support_tickets_raised',
    'satisfaction_score'
]

outliers_dict = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outliers_dict[col] = len(outliers)
    print(f"{col}: {len(outliers)} outliers")
```
